The actual code for Auto-encoding

```python
import torch
import torch.nn as nn
import pandas as pd
from torch.utils.data import DataLoader, TensorDataset

df = pd.read_csv("./Biked_Reference_data/scaled_features.csv")
```

```python
df.drop(columns=["Unnamed: 0.1", "Unnamed: 0"], inplace=True, errors='ignore')
```

```python
df
```

```python
print("--- DATA DIAGNOSTICS ---")

# 1. Check for NaNs and Infinite values (These will instantly break training)
print(f"1. Missing (NaN) values: {df.isna().sum().sum()}")
print(f"2. Infinite values: {np.isinf(df.select_dtypes(include=np.number)).values.sum()}")

# 2. Check for Zero Variance (Constant Features)
# If a feature is exactly the same for every bike, its standard deviation is 0.
# StandardScaler handles this poorly, and it's useless for the model.
zero_var_cols = df.columns[df.std() == 0]
print(f"3. Constant columns (0 variance): {len(zero_var_cols)}")
if len(zero_var_cols) > 0:
    print(f"   Drop these columns: {list(zero_var_cols)}")

# 3. Check the Scaling (Did StandardScaler work?)
# Mean should be practically 0, Std Dev should be practically 1
print(f"4. Average Mean across all features: {df.mean().mean():.4f} (Should be ~0)")
print(f"5. Average StdDev across all features: {df.std().mean():.4f} (Should be ~1)")

# 4. The Outlier Check (Crucial for MSE Loss)
# In a standard normal distribution, 99.7% of data falls between -3 and 3.
# Anything beyond ±5 is a severe outlier. Anything beyond ±10 is a red flag.
max_val = df.max().max()
min_val = df.min().min()
print(f"6. Absolute Maximum value: {max_val:.4f}")
print(f"7. Absolute Minimum value: {min_val:.4f}")

# Find how many severe outliers exist
outlier_mask = (df > 5) | (df < -5)
print(f"8. Number of severe outliers (>5 or <-5): {outlier_mask.sum().sum()}")
```

```python
class BikeAutoencoder(nn.Module):
    def __init__(self):
        super(BikeAutoencoder, self).__init__()
        
        # 1. THE ENCODER: Compresses 96 parameters down to 10
        self.encoder = nn.Sequential(
            nn.Linear(83, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 10) # The Latent Space (Your 10 Sliders)
        )
        
        # 2. THE DECODER: Expands 10 sliders back to 96 parameters
        self.decoder = nn.Sequential(
            nn.Linear(10, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.Linear(128, 83)
            # Use Sigmoid if your data is normalized between 0 and 1
        )

    def forward(self, x):
        latent_space = self.encoder(x)
        reconstructed_bike = self.decoder(latent_space)
        return reconstructed_bike
```

```python
data_tensor = torch.tensor(df.values, dtype=torch.float32)
dataset = TensorDataset(data_tensor, data_tensor)
dataloader = DataLoader(dataset, batch_size=128, shuffle=True)
```

```python
model = BikeAutoencoder() # (The class from the previous message, minus the Sigmoid)
criterion = nn.MSELoss()  # Mean Squared Error
optimizer = torch.optim.Adam(model.parameters(), lr=0.0005)

epochs = 200

print("Starting training...")
for epoch in range(epochs):
    for batch_data, _ in dataloader:
        
        # 1. Forward pass: Compress and decompress
        reconstructed = model(batch_data)
        
        # 2. Calculate the error
        loss = criterion(reconstructed, batch_data)
        
        # 3. Backward pass: Update the weights
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    if (epoch+1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

print("Training complete! Model has learned to build bikes.")
```
